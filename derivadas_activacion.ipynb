Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5,num=N)
```

Derivada Discreta

```python
def df(f):
  h=0.00001
  return (f(x+h)-f(x))/h
```

Función Lineal

```python
def f(x):
  return x
  
plt.plot(x, f(x), 'b')
plt.plot(x, df(f),'r')
plt.grid()
```

Función Heaviside

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y

N = 1000

plt.plot(x, H(x), 'b')
plt.plot(x, df(H),'r')
plt.grid()
```

Función sigmoide

```python
def f(x):
  return 1/(1+ np.exp(-x))

N = 1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x, df(f), 'r')
plt.grid()
```

Función tangente hiperbólica

```python
def f(x):
  return np.tanh(x)
N = 1000
y = f(x)
plt.plot(x, f(x), 'b')
plt.plot(x,df(f),'r')
plt.grid()
```

Función ReLU

```python
def f(x):
  return np.maximum(x,0)
N = 1000
y = f(x)
plt.plot(x,f(x),'b')
plt.plot(x,df(f),'r')
plt.grid()
```
